# Mandelbrot Set

```python
def plane(xmin, xmax, ymin, ymax, pixel_density): # returns Argand plane
    re = np.linspace(xmin, xmax, int((xmax-xmin)*pixel_density))
    im = np.linspace(ymin, ymax, int((ymax-ymin)*pixel_density))
    return re[np.newaxis, :] + 1j*im[:, np.newaxis]

def mandelbrot(c, n_iter): # returns array of boolean: True if c is in Mandelbrot set and False otherwise
    z = 0
    for _ in range(n_iter):
        z = z**2 + c
    return abs(z) <= 2
```

```python
c = plane(-1.5, 0.5, -1, 1, 512)
elem = c[mandelbrot(c, 20)] #filtering: returns only elements in Mandlebrot set
plt.scatter(elem.real, elem.imag, marker=',', s=1)
```

```python
plt.imshow(mandelbrot(c, 20), cmap="binary")
```
